# Short sample problem 1:  Loop (5 min)

Write a function that returns the next power of 2 after the argument (or the argument, if it is a power of 2).  For example, next_power(5) == 8, and next_power(8) == 8 as well.  You can assume the argument is at least 1.

```python
def next_power(my_int):
  min_power_of_2 = 1
  while my_int > min_power_of_2:
    min_power_of_2 = min_power_of_2 * 2
  return min_power_of_2

print(next_power(5)) # Expect 8
print(next_power(8)) # Expect 8
print(next_power(32)) # Expect 32
print(next_power(33)) # Expect 64
```
