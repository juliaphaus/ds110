You are considering baby names for your children - you expect to have two.  The names you've narrowed it down to are:

a, 5 pts) Write a loop that prints out all possible combinations of two baby names, one per line, such as "Zephyrus-Umaira".  Do not repeat the same name on the same line.  You can repeat the same combination in a different order, however, since the first name will be the name of the older child.


````python
babynames = [
    "Zephyrus", "Clementine", "Hendrixon", "Quintavius", "Maverick", "Xanthippe",
    "Bellatrix", "Orionis", "Jupiter", "Vespera", "Djangolina", "Ravenwood",
    "Echolette", "Sableine", "Peregrine", "Wrenly", "Ivyanna", "Lunaria",
    "Fitzgerald", "Gizmonda", "Axelrod", "Nashira", "Yaralyn", "Talonis",
    "Kodaline", "Umaira"
]

for first in babynames:
  for second in babynames:
    if first != second:
      print(str(first) + "-" + str(second))
````


b, 5 pts) Well, that's too many combinations to consider.  You decide to limit the search to names that have at least two "e"'s to make it sound more mellifluous.  Use a for loop to create a list of such names.  You can use the `.count()` string method to figure out whether a name has enough e's (s.count(n) returns the number of times n appears in s).  Check for both uppercase and lowercase.  You don't need to do the nested loop again - just print the short list of two-e names.  (Be sure it's also assigned to a variable so that you can do the next part.)


````python

for first in babynames:
  for second in babynames:
    if first != second:
      if (first.lower()).count("e") >= 2:
        if (second.lower()).count("e") >= 2:
          print(str(first) + "-" + str(second))
````

c, 8 pts) Wait - middle names!  Print all combnations of two different names followed by two other different names, using your shorter list of names that have enough e's.  For example, "Xanthippe Yaralyn-Xanthippe Umaira" refers to two sisters who both have a first name of Xanthippe.  As this example suggests, you can now repeat first names between individuals, since we have the middle names to distinguish them, but no individual should get a middle name that is the same as their first name (no "Xanthippe Xanthippe") and the two individuals should not have identical name combinations (no "Xanthippe Yaralyn-Xanthippe Yaralyn").  Tip:  You might find it useful to write the first draft of the code without quite so many rules.


````python
for first in babynames:
  for second in babynames:
    for third in babynames:
      for fourth in babynames:
        if first != second:
          if third != fourth:
            if (first, second) != (third, fourth):
              if (first.lower()).count("e") >= 2:
                if (second.lower()).count("e") >= 2:
                  if (third.lower()).count("e") >= 2:
                    if (fourth.lower()).count("e") >= 2:
                      print(str(first) + " " + str(second) + "-" + str(third) + " " + str(fourth))
````


d, 7 pts) How are you going to decide between all these combinations?  Well, you still like the letter 'e', you decide.  Copy your code from the last section to iterate through all the combinations of two names with two other names (again just using names with at least two e's), but this time, keep track of which combination has the most e's (uppercase or lowercase) between all four names.  Print the number of e's at the end of every line, and finally print "Best names: " followed by the combination with the most e's.  In case of a tie, you only need to print one of the best combinations.

````python
best_count = 0
best_combo = ""
for first in babynames:
  for second in babynames:
    for third in babynames:
      for fourth in babynames:
        if (first.lower()).count("e") >= 2 and (second.lower()).count("e") >= 2 and (third.lower()).count("e") >= 2 and (fourth.lower()).count("e") >= 2:
          if first != second and third != fourth:
            if (first, second) != (third, fourth):
              e_count_1 = (first.lower()).count("e")
              e_count_2 = (second.lower()).count("e")
              e_count_3 = (third.lower()).count("e")
              e_count_4 = (fourth.lower()).count("e")
              e_count_total = e_count_1 + e_count_2 + e_count_3 + e_count_4

              #combos = first + "" + second + "-" + third + " " + fourth
              if e_count_total > best_count:
                best_count = e_count_total
                best_combo = str(first) + " " + str(second) + " - " + str(third) + " " + str(fourth)
print("Best names: " + best_combo)
print(str(best_count) + " e's")
````
