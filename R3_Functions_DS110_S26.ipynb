**for loop** - a control flow statement that repeatedly executes a block of code until a condition is met. Used when the number of iterations is known. 

**Tuple** - an immutable data structure that holds multiple pieces of data.

The `zip(list1, list2)` function takes as input two lists, list1 and list2, and aggregates the elements in these lists into tuples. The function returns an iterable object containing these tuples. To understand this concept run the code below.

```python
letters = ['a', 'b', 'c', 'd', 'e']
numbers = [1,2,3,4,5]
zipped = zip(letters, numbers)
print(zipped) # an object, not a list zipped[0] will fail
for item in zipped: # we can loop over zipped, that's what being iterable means
    print(item) # each item in zipped is a tuple containing ("a", 1), ("b", 2), etc
```

# Tool focus:  doctest

A classy way to write tests is to use doctest, which uses test cases that you provide in comments with the function.  (You aren't expected to use this for homework or exams.)  Put the test examples in the "docstring" within the function, as shown here, then run that cell with the docstring inside, and then call `doctest.testmod()` to check that they all work as expected.  (If no error is reported, it's fine. You can try messing up the test and see what happens in that case.)

```python
import doctest
import statistics

def min_mean_max(lst):
    """
    Return the min, mean, and max value in lst as 3 return values.
    If the list is empty, return (0, 0, 0) instead.
    
    >>> min_mean_max([1,2,3,4,5])
    (1, 3, 5)
    
    >>> min_mean_max([])
    (0, 0, 0)
    """
    if not lst:  # meaning it's empty - some Python slickness
        return (0,0,0)
    return min(lst), statistics.mean(lst), max(lst)
```

```python
doctest.testmod()
```
